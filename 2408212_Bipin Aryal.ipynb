```python
#Task 1 - Setup Task - Preparing the South-Asia Dataset:
import pandas as pd

south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
df = pd.read_csv('/content/drive/MyDrive/Bipin/Assignment/WHR-2024-5CS037.csv')
south_asia_df = df[df['Country name'].isin(south_asian_countries)]
south_asia_df.to_csv('south_asia_happiness.csv', index=False)
print(south_asia_df)
```

```python
#Task - 2 - Composite Score Ranking

# Composite Score calculation
south_asia_df['Composite Score'] = (0.40 * south_asia_df['Log GDP per capita'] +0.30 * south_asia_df['Social support'] + 0.30 * south_asia_df['Healthy life expectancy'])

# Rank of countries based on Composite Score in descending order
south_asia_df = south_asia_df.sort_values(by='Composite Score', ascending=False)
south_asia_df = south_asia_df.reset_index(drop=True)

# Visualization of top 5 countries using horizontal bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Composite Score', y='Country name', data=south_asia_df.head(5), palette="viridis")
plt.title('Top 5 South Asian Countries by Composite Score')
plt.xlabel('Composite Score')
plt.ylabel('Country')
plt.show()


# Comparing rank with original Score
plt.figure(figsize=(12, 6))
plt.plot(south_asia_df['score'], color='red', label='Original Score')
plt.plot(south_asia_df['Composite Score'], color='blue', label='Composite Score')
plt.xticks(range(len(south_asia_df)), south_asia_df['Country name'], rotation=90)

plt.title('Comparison of Original Score and Composite Score in South Asia')
plt.xlabel('Country')
plt.ylabel('Score')
plt.legend()
plt.show()

#discuss part;
print("Correlation : In scatter plot, the points in line with 45 degrees are considered having good validity of the original score assessment with the composite score. Omissions and variations are marked next to the line.")
print("Rank Alignment:One of the elements that are essential for analysis in a bar plot is whether the existence of the shift between the original and the composite score is substantial. These substantive differences tell us that the composite scoring might measure factors in a different manner.")
print("Insights:Discuss why some countries have deviation. For instance, a country may be richer measured in terms of GDP but is less happier than a country which is nearer to the index; thus such a country would receive a lower composite rank.These findings are useful in establishing whether the two rankings correlate, and setting the stage for explaining disparities in the observed rankings.")
```

```python
#Task - 3 - Outlier Detection:
#1. Calculating IQR for GDP per Capita and Score
Q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

#2. Define outlier boundaries
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Identifing outliers
outliers = south_asia_df[
    (south_asia_df['Log GDP per capita'] < lower_bound_gdp) |
    (south_asia_df['Log GDP per capita'] > upper_bound_gdp) |
    (south_asia_df['score'] < lower_bound_score) |
    (south_asia_df['score'] > upper_bound_score)
]

# 3.Create the scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, hue=south_asia_df.index.isin(outliers.index), palette={True: 'green', False: 'red'})
plt.title('Outlier Detection in South Asian Countries')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
print("Outlier Countries:")
print(outliers[['Country name', 'Log GDP per capita', 'score']])
```

**Characteristics of the Outliers**
Extremes in GDP or Score:
Outliers identified using the IQR method for either "Log GDP per capita" or "Happiness Score" would be well above or way below the normal rangethat is, 1.5 times the IQR. That could include countries with

Exceptionally high or low GDP per capita compared to regional averages.
Unusually high or low levels of happiness scores, reflecting unique societal, political, or economic circumstances.
Possible Reasons:

High Outliers:
Its possible the more well-heeled countries have won on the back of unequal socio-economic factors like an explosion of resource wealth and tech industries.
High scores may indicate good social systems, effective governance, or high levels of individual freedoms.
Low Outliers:
Low GDP outliers may face economic instability, resource scarcity, or any governance challenges. The low scores could be indicative of such issues as conflict, inequality, or poor access to basic services.

Geography or Political Context: Some countries might be outliers due to geographical isolation, size, or political policies drastically affecting economic output or social wellbeing.

**Potential to Affect the Regional Averages**
Biased Averages:

In this respect, the presence of outliers may be highly disproportional towards regional averages in GDP and scores of happiness which then make the measures non-representative of the mass.
Policy Misinterpretations:

Outliers, if not accounted for, may give a wrong reflection of regional trends. An extremely rich country, for instance, could suggest that a region has a higher GDP when in fact most countries' experience is different. Bias in Comparative Analysis:

These outliers make the South Asia averages a little hard to compare with other regions because their removal or better accounting would better represent the means:.

Implications for Interventions: Such outliers may help policymakers to zero in on abnormalities, whether to scale up the success of high outliers or address the problems of low outliers.





```python
# Task - 4 - Exploring Trends Across Metrics

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

south_asia_df = pd.read_csv('south_asia_happiness.csv')

# Pearson Correlation Calculation
metrics = ['Freedom to make life choices', 'Generosity']
for metric in metrics:
    correlation = south_asia_df['score'].corr(south_asia_df[metric], method='pearson')
    print(f"Pearson correlation between Score and {metric}: {correlation}")

# Scatter plots with trendlines
for metric in metrics:
    plt.figure(figsize=(10, 8))
    sns.regplot(x=metric, y='score',color='blue', data=south_asia_df)
    plt.title(f'Scatter Plot of Score vs. {metric} in South Asian Countries')
    plt.xlabel(metric)
    plt.ylabel('Score')
    plt.show()
```

```python
#Task 5 - Gap Analysis:

# Add GDP-Score Gap column
south_asia_df['GDP-Score Gap'] = south_asia_df['Log GDP per capita'] - south_asia_df['score']

# Rank by GDP-Score Gap
south_asia_df_ascending = south_asia_df.sort_values(by='GDP-Score Gap')
south_asia_df_descending = south_asia_df.sort_values(by='GDP-Score Gap', ascending=False)

# Highlight top 3 countries with largest positive and negative gaps
plt.figure(figsize=(10, 8))

# Top 3 positive gaps
plt.subplot(1, 2, 1)
sns.barplot(x='Country name', y='GDP-Score Gap', data=south_asia_df_descending.head(3), palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title('Top 3 Countries-Largest Positive GDP-Score Gap')

# Top 3 negative gaps
plt.subplot(1, 2, 2)
sns.barplot(x='Country name', y='GDP-Score Gap', data=south_asia_df_ascending.head(3), palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title('Top 3 Countries-Largest Negative GDP-Score Gap')

plt.tight_layout()
plt.show()
```

```python
# Task 2: Top and Bottom Performers

def plot_top_bottom(df, region_name):

    region_df_sorted = df.sort_values(by='score', ascending=False)
    top3 = region_df_sorted.head(3)
    bottom3 = region_df_sorted.tail(3)
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.barplot(x='Country name', y='score', data=top3, palette="viridis")
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Top 3 Happiest Countries in {region_name}')

    plt.subplot(1, 2, 2)
    sns.barplot(x='Country name', y='score', data=bottom3, palette="viridis")
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Bottom 3 Happiest Countries in {region_name}')

    plt.tight_layout()
    plt.show()

    plot_top_bottom(middle_east_df, "Middle East")
    plot_top_bottom(south_asia_df, "South Asia")
```
